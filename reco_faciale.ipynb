```python
# Cropping des visages
# On récupère le premier visage détecté
top, right, bottom, left = face_locations[0]
# On découpe l'image en utilisant les coordonnées du visage
face_image = image[top:bottom, left:right]
# On affiche le visage découpé
plt.imshow(face_image)
plt.axis('off')
plt.show()
```

```python
face_locations_unknown = face_recognition.face_locations(unknown_image)
face_locations_unknown
```

```python
# On récupère le deuxième visage détecté
top, right, bottom, left = face_locations_unknown[0]
# On découpe l'image en utilisant les coordonnées du visage
face_image_unknown = unknown_image[top:bottom, left:right]
# On affiche le visage découpé
plt.imshow(face_image_unknown)
plt.axis('off')
plt.show()
```

```python
# Encoding avec les images complètes
known_encoding = face_recognition.face_encodings(known_image)[0]
unknown_encoding = face_recognition.face_encodings(unknown_image)[0]
```

```python
# Essai d'encoding avec le paramètre known_face_locations de la méthode face_encodings
# Encoding avec les visages découpés  (code source https://github.com/ageitgey/face_recognition/blob/master/face_recognition/api.py#L203)
k_encoding = face_recognition.face_encodings(image, known_face_locations=face_locations)[0]
```

```python
# Comparaison de 2 photos différentes d'une même personne
brad_beard = face_recognition.load_image_file("src/images/brad_beard.jpg")
brad_short_hair = face_recognition.load_image_file("src/images/brad_short_hair.jpg")
```

```python
plt.imshow(brad_beard)
plt.axis('off')
```

```python
plt.imshow(brad_short_hair)
plt.axis('off')
```

```python
beard_encoding = face_recognition.face_encodings(brad_beard)[0]
short_hair_encoding = face_recognition.face_encodings(brad_short_hair)[0]
```

```python
results = face_recognition.compare_faces([beard_encoding], short_hair_encoding)
results
```

```python
# Comparaison avec lunettes
brad_glasses = face_recognition.load_image_file("src/images/brad_glasses.jpg")
```

```python
brad_glasses_encoding = face_recognition.face_encodings(brad_glasses)[0]
```

```python
results = face_recognition.compare_faces([beard_encoding], brad_glasses_encoding)
results
```

```python
# Distance entre les visages
face_distances = face_recognition.face_distance([beard_encoding], short_hair_encoding)
```
